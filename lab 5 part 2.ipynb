```python
sns.set_style("whitegrid");

sns.FacetGrid(df_with_delimiter, hue="target_class", height=10).map(plt.scatter,
                                    "sepal length(cm)", "sepal width (cm)").add_legend();

plt.show();
```

```python
sns.set_style("whitegrid");

sns.pairplot(df_with_delimiter,vars =['sepal length(cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)','target_class']
             , hue = 'target_class', diag_kind = 'kde',plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'}, height = 2) 

plt.show()
```
